```python
compradores['first_orders_dt']=pd.to_datetime(compradores['first_orders_dt'])
compradores['primer_sesion_dia']=pd.to_datetime(compradores['primer_sesion_dia'])
compradores.head()
```

```python
compradores['dias_primera_compra']=((compradores['first_orders_ts'] - compradores['primera_sesion_ts'])/np.timedelta64(1,'D')).astype('int')
compradores.head()
```

```python
compradores['dias_primera_compra'].plot(kind='hist')
```

$2.2$¬øCu√°ntos pedidos hacen durante un per√≠odo de tiempo dado?

```python
cohort_size=compradores.groupby('first_orders_month').agg({'Uid':'nunique'}).reset_index()
cohort_size.rename(columns={'Uid':'n_buyers'},inplace=True)
cohort_size.head()
```

```python
#cohortsmerge
cohort=pd.merge(orders,compradores)
cohort.head()
```

```python
cohort=cohort.groupby(['first_orders_month', 'order_month']).agg({'Revenue':'count'}).reset_index()
cohort.rename(columns={'Revenue':'n_orders'},inplace=True)
```

```python
cohort['ciclo_de_vida']=(cohort['order_month'] - cohort['first_orders_month']).apply(lambda x:x.n)
cohort.head()
```

```python
report_cohort=pd.merge(cohort_size,cohort)
report_cohort.head()
```

```python
report_cohort['orders_buyer']=report_cohort['n_orders']/report_cohort['n_buyers']
report_cohort.head()
```

```python
analisis_cohort_pivot=report_cohort.pivot_table(
    index='first_orders_month',
    columns='ciclo_de_vida',
    values='orders_buyer',
    aggfunc='sum'
).cumsum(axis=1)
analisis_cohort_pivot.round(2).fillna("")
```

2.4¬øCu√°nto dinero traen? ($LTV$)

```python
# F√≥rmula simple de LTV = ingreso_promedio * frecuencia * tiempo_de_vida
#orders['LTV'] = df['promedio de compras por mes'] * df['frecuencia'] * df['tiempo_de_vida']
```

```python
primer_visit.head()
```

```python
orders.head()
```

```python
revenue_users=orders.groupby('Uid')['Revenue'].sum().reset_index(name='total_revenue')
revenue_users
```

```python
# Unir ordenes con la cohorte de cada usuario
orders_cohorte = pd.merge(revenue_users, primer_visit[['Uid', 'primer_sesion_mes']], on='Uid')

orders_cohorte.head()
```

```python
#agrupacion LTV POR COHORTE en pri
df_ltv=orders_cohorte.groupby('primer_sesion_mes').agg(
                total_revenue=('total_revenue', 'sum'),
                total_users=('Uid', 'nunique'),
            ).reset_index()
```

```python
df_ltv
```

```python
#proporcionLTV cuanto dinero aporta cada usuario que inicio secion por usuario cada mes por primera ves
df_ltv['ltv']=df_ltv['total_revenue']/df_ltv['total_users']
df_ltv=df_ltv.sort_values(by='primer_sesion_mes')
df_ltv
```

```python
df_ltv.set_index('primer_sesion_mes')['ltv'].plot(
    kind='bar',
    title='LTV por cohorte mensual ',
    ylabel='LTV promedio por usuario',
    xlabel='Mes de primera sesi√≥n',
    figsize=(10, 5),
    color='skyblue'
)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
cohort_size=compradores.groupby('first_orders_month').agg({'Uid':'nunique'}).reset_index()
cohort_size.rename(columns={'Uid':'n_buyers'},inplace=True)
cohort_size.head()
```

```python
cohorts_=pd.merge(orders,compradores)
```

```python
cohorts_=cohorts_.groupby(['first_orders_month','order_month']).agg({'Revenue':'sum'}).reset_index()
```

```python
cohorts_['ciclo_de_vida']=(cohorts_['order_month'] - cohorts_['first_orders_month']).apply(lambda x:x.n)
cohorts_.head()
```

```python
cohorts_ltv=pd.merge(cohort_size, cohorts_)
cohorts_ltv['ltv']=cohorts_ltv['Revenue']/cohorts_ltv['n_buyers']
cohorts_ltv.head()
```

```python
LTV=cohorts_ltv.pivot_table(
    index='first_orders_month',
    columns='ciclo_de_vida',
    values='ltv',
    aggfunc='sum'
).cumsum(axis=1)
LTV.round(2).fillna('')

```

```python
#dinero que traen LTV
sns.heatmap(LTV,annot=True)
plt.show()
```

```python

for source,df in compradores.groupby('Source Id'):
    if source !=7:
        cohort_size=(
            df.groupby(['Source Id','first_orders_month'])
            .agg({'Uid':'nunique'})
            .reset_index()
            
        )
        cohort_size.columns=['source_id','first_orders_month','n_buyers']
        cohort_revenue=(
            df.groupby(['Source Id','first_orders_month','order_month'])
            .agg({'Revenue':'sum'})
            .reset_index()
        )
        cohort_revenue.columns=['source_id','first_orders_month','order_month','gp']

        report_romi=pd.merge(cohort_size,cohort_revenue,on=['source_id','first_orders_month'])
        report_romi['ciclo_de_vida']=(report_romi['order_month'] - report_romi['first_orders_month']).apply(lambda x:x.n)
        report_romi['ciclo_de_vida']=report_romi['ciclo_de_vida'].round().astype('int')
        report_romi['ltv']=report_romi['gp']/report_romi['n_buyers']
        report_romi=pd.merge(report_romi,costs.groupby(['source_id','cost_month']).agg({'costs':'sum'}).reset_index(),
                             left_on=['source_id','first_orders_month'],right_on=['source_id','cost_month'])
        report_romi=report_romi.drop(['cost_month'],axis=1)
        report_romi['cac']=report_romi['costs']/report_romi['n_buyers']
        report_romi['romi']=report_romi['ltv']/report_romi['cac']

        romi=report_romi.pivot_table(
            index='first_orders_month',
            columns='ciclo_de_vida',
            values='romi',
            aggfunc='mean'
        ).cumsum(axis=1)
        romi.round(2).fillna('')
        
        romi.index=romi.index.astype(str)
        sns.heatmap(romi,annot=True,fmt='.2f',linewidths=1).set(title='romi_por_fuente{}'.format(source))
        plt.show()
```
