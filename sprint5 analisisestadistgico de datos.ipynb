```python
def calcular_tarifas(data):
    tarifa_base=data['usd_monthly_pay']
    
    #calcular llamadas
    exceso_min=data['total_minutos'] - data['minutes_included']
    exceso_min2=max(0,exceso_min)
    minutos_ingreso= data['usd_per_minute']*exceso_min2
    minutos_ingreso
    
     #calcular mensages
    exceso_message=data['total_de_mensajes_count'] - data['messages_included']
    exceso_message2=max(0,exceso_message)
    message_ingreso= data['usd_per_message']*exceso_message2
    message_ingreso
    
    
    #calcular internet
    exceso_internet=data['total_de_internet_sum'] - data['gb_per_month_included']
    exceso_internet2=max(0,exceso_internet)
    internet_ingreso= data['usd_per_gb']*exceso_internet2
    internet_ingreso
    
    
    #calcular_la_tarifa_final
    ingresos_totales= tarifa_base+minutos_ingreso+message_ingreso+internet_ingreso
    return ingresos_totales
```

```python
merged_4['ingresos_totales']=merged_4.apply(calcular_tarifas, axis=1)

```

```python
# Compara la duración promedio de llamadas por cada plan y por cada mes. Traza un gráfico de barras para visualizarla.
# Asumiendo que 'merged_4' ya está definido
promedio_de_llamadas = merged_4.groupby(['month', 'plan']).mean()['total_minutos'].reset_index()

# Obtener los meses únicos y los planes únicos para la estructura de la gráfica
meses = promedio_de_llamadas['month'].unique()
planes = promedio_de_llamadas['plan'].unique()

# Ancho de cada barra
ancho_barra = 0.35

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(12, 7))

# Iterar sobre cada plan y crear las barras correspondientes
for i, plan in enumerate(planes):
    datos_plan = promedio_de_llamadas[promedio_de_llamadas['plan'] == plan]
    # Calcular las posiciones de las barras para este plan, desplazándolas ligeramente
    posiciones = np.arange(len(meses)) + i * ancho_barra
    ax.bar(posiciones, datos_plan['total_minutos'], ancho_barra, label=plan)

# Establecer las etiquetas del eje X en el centro de los grupos de barras
ax.set_xticks(np.arange(len(meses)) + ancho_barra / 2)
ax.set_xticklabels(meses)
plt.xticks(rotation=45, ha='right')

# Añadir etiquetas a los ejes y título
ax.set_xlabel("Mes")
ax.set_ylabel("Duración Promedio (minutos)")
ax.set_title("Comparación de la Duración Promedio de Llamadas por Plan y Mes")

# Añadir la leyenda para distinguir los planes
ax.legend(title='Plan')

# Ajustar el layout
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```
